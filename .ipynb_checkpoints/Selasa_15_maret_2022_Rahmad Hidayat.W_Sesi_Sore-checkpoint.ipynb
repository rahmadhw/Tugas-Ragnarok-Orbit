```python
cetak = ['Age', 'Gender']
for cetak1 in cetak:
    plt.figure(figsize=(20,10))
    plt.title(f"{cetak1}")
    sns.countplot(df[cetak1])
    plt.show()
```

### Numerik vs Target

```python
plt.figure(figsize=(7, 6))
sns.distplot(df.Age[df.Exited == 0], bins=[0, 5, 12, 18, 40, 120], color='r', label="Keluar")
sns.distplot(df.Age[df.Exited == 1], bins=[0, 5, 12, 18, 40, 120], color='g', label="Masuk")
plt.legend();
```

### Lebih Cendurung Orang yang berumur <= 40 ,berdasarkan data lebih tinggi dri pada orang yang berumur > 40 tahun

```python
font = {'size':15, 'family':'normal'}
plt.rc('font', **font)
plt.figure(figsize=(20,12))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
```
