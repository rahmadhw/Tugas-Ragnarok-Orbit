## gender vs class exited

```python
df.head()
```

```python
data_gender = df["Gender"]
data_gender
```

```python
data_exited = df['Exited']
data_exited
```

```python
# Set up the matplotlib figure
f, axes = plt.subplots(ncols = 1,figsize=(5, 5), sharex=True)
sns.despine(left=True)  #gender-vs-class exited

# Ploting the histogram with KDE

sns.histplot(data=data_gender,label="gender",color='b',kde=True,linewidth=2,alpha=0.3)
sns.histplot(data=data_exited,label="exited",kde=True,color='orange',linewidth=2,alpha=0.3)


plt.legend()
plt.tight_layout()
```

##### Berdasarkan data Di ataas perbandingan antara gender mana yang lebih sering keluar atau exited yaitu yang bergender Wanita
